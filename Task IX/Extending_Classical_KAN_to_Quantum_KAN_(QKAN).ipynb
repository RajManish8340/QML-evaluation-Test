# Extending Classical KAN to Quantum KAN (QKAN)

## Introduction

- **Kolmogorov-Arnold Networks (KAN)** are neural networks that use flexible, adaptive basis functions instead of fixed activations.
- **Quantum Kolmogorov-Arnold Networks (QKAN)** leverage quantum circuits to represent functions, potentially improving computational efficiency and modeling complex data.

## Why Extend to Quantum KAN?

- Quantum systems offer **exponential computational advantages** in certain tasks.
- **High-dimensional data** can be better represented using quantum states.
- Quantum circuits can provide more efficient representations of functions using fewer parameters.


## Classical KAN Overview

- Uses adaptive, learnable functions instead of fixed activation functions.
- Employs a small number of hidden layers with complex, expressive nodes.
- Effective in tasks like regression and symbolic regression.

## Quantum KAN Concept

- **Quantum Nodes**: Each node in the network represents a quantum circuit.
- **Parameterized Quantum Circuits (PQCs)**: quantum circuits with trainable parameters serve as adaptable components in QKAN.
- **Quantum Measurement**: Extracts the output for further processing.
- Can leverage algorithms like **VQE (Variational Quantum Eigensolver)** for optimization.



## 1. Input Layer

### Purpose
- The input layer converts classical data into quantum data, allowing it to be processed by a quantum circuit.
- This is done using quantum encoding techniques.

### Quantum Encoding Techniques
Two primary encoding methods are commonly used:

1. **Amplitude Encoding:**
    - Maps the input data to the amplitudes of a quantum state.
    - Efficient for high-dimensional data.
    - Example: A vector $$ [x_1, x_2, …, x_n] $$ is encoded as a quantum state $$ |\psi\rangle = \sum c_i |i\rangle $$.

2. **Angle Encoding:**
    - Encodes classical data using the rotation angles of quantum gates.
    - Each data point is represented as a rotation on a qubit.
    - Example:
     $$ R_y(x_i) |0\rangle = \cos(x_i/2)|0\rangle + \sin(x_i/2)|1\rangle $$.

The choice of encoding depends on factors such as data dimensionality and the type of problem.

## 2. Hidden Layers

### Purpose
- The hidden layers consist of quantum nodes that apply transformations using parameterized quantum circuits (PQCs).
- The PQCs act as the quantum equivalent of neural network layers.

### Structure
Each hidden layer typically includes:

1. **Quantum Nodes:**
    - Nodes in this layer represent qubits manipulated by quantum gates.
2. **Parameterized Quantum Circuits (PQCs):**
    - Variational quantum circuits with learnable parameters.
    - Common gates include:
      - **RZ (Z-axis rotation)**: Rotates a qubit around the Z-axis.
      - **RX (X-axis rotation)**: Rotates a qubit around the X-axis.
      - **CNOT (Controlled-NOT)**: Entangles qubits, allowing interactions.
3. **Variational Parameters:**
    - Quantum gates contain parameters that are updated during training.
    - The optimization process minimizes a loss function based on the measurement outcomes.

### Quantum Circuit Example
- A single hidden layer might consist of:
  - **Rotation Gates (RZ, RX)** for applying transformations.
  - **Entanglement Gates (CNOT)** to create correlations between qubits.
  - **Measurement Gates** to extract values from the circuit.

---

## 3. Measurement Layer

### Purpose
- The final quantum state needs to be converted into a classical output.
- Measurements collapse the quantum state to produce a set of classical values.

### Process
- After applying the PQCs, the qubits are measured using computational basis states $$ e.g., |0\rangle and |1\rangle $$.
- The measurement results represent probabilities based on the qubit states.
- Multiple measurements are typically performed to obtain statistically accurate results.

---

## 4. Output Layer

### Purpose
- The output layer interprets the measured values to generate a final prediction.
- This often involves applying a simple classical function to map the results to meaningful outputs.

### Possible Outputs
- **Regression Problems:** The measured values can represent continuous outputs.
- **Classification Problems:** The values can be processed using a softmax function to determine the most likely class.

### Example
- For a binary classification task, measurements from the qubits can be passed through a sigmoid function to obtain a probability score.

$$ P(y=1 | x) = \sigma(\text{Measured Value}) = \frac{1}{1 + e^{-x}} $$

## Conclusion

The proposed QKAN architecture leverages the power of quantum computing to enhance classical neural networks. By using quantum encoding, parameterized quantum circuits, and measurement-based output generation, QKANs can tackle complex problems with greater efficiency. This makes them particularly suitable for high-dimensional data analysis, such as HEP experiments.

Future improvements could involve optimizing the choice of quantum gates, designing better encoding schemes, and scaling the architecture to larger datasets using quantum hardware advancements.

---


## Advantages of QKAN

- Improved model expressiveness with fewer parameters.
- Faster training and inference on quantum hardware.
- Enhanced performance on tasks like classification and symbolic regression.
