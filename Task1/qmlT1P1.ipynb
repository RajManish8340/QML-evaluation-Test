## impoting neccessary libraries

```python
import pennylane as qml
from pennylane import numpy as np
import matplotlib.pyplot as plt
```

# quantum device with 5 qubits

```python
dev = qml.device("default.qubit", wires=5)
```

# Simple Quantum Operations

```python
@qml.qnode(dev) # This decorator converts the function into a quantum node (QNode) that runs on the specified device
def quantum_circuit():

    # Apply Hadamard gate to all qubits to create superposition
    for i in range(5):
        qml.Hadamard(wires=i)

    # Apply CNOT gate to create entaglement
    cnot_pairs = [(0, 1), (1, 2), (2, 3), (3, 4)] # Define CNOT connections
    for control, target in cnot_pairs:
        qml.CNOT(wires=[control, target])

    # SWAP gate between qubits 0 and 4
    qml.SWAP(wires=[0, 4])

    # Apply Rx(pi/2) on qubit 2
    qml.RX(np.pi / 2, wires=2)

    return qml.state()
```

```python
fig, ax = qml.draw_mpl(quantum_circuit)()
plt.show()
```
